## Objective

The aim of this demo is to show how to visualize neural network architectures. As an example I have selected a ready model `Xception` found from keras applications.

## Text format summary

`model.summary()` shows the text formatted list of the layers, their types, output shapes, number of parameters, and how the layers are connected to each other. However, in this presentation it is difficult to see the branches or residual connections that this example, `Xception` has.

```python
from keras.applications.xception import Xception

# Show the summary of Xception model (in text format)
base_model = Xception(input_shape = (600, 600, 3))
base_model.summary()
```

## Graphical presentation

Next we study the same model using graphical presentation. The model structure is stored to png-file and the displayed in a Markdown cell. The image shown below is created using Markdown image formatting syntax. For more details, see for example: [Markdown Cheatsheet](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet)

```python
# Save the model to image
# See: https://keras.io/visualization/
from keras.utils import plot_model
plot_model(base_model, to_file='xception.png', show_shapes = True)
```
